```python
Roster=['G_p', 'G_c', 'G_1b', 'G_2b','G_3b', 'G_ss', 'G_lf', 'G_cf', 'G_rf']
```

```python
Roster
```

```python
bfdata=bfdata.replace([np.inf, -np.inf], np.nan).dropna(subset=["obpsal"], how="all")
```

```python
bfdata=bfdata.head(5000)
```

```python
def mb_players(year, pos):
    return bfdata[(bfdata['yearID'] == year) & (bfdata[pos] > 20)].sort_values('obpsal', ascending=False)
```

```python
G_p=[]
```

```python
G_c=[]
```

```python
G_rf=[]
```

```python
def work(pos):
    r=mb_players(2001,pos)
    m=r.loc[r[pos].idxmax()]
    return m
```

```python
for pos in Roster:
    v=work(pos)
    if pos=='G_p':
        G_p=v
    if pos=='G_c':
        G_c=v
    if pos=='G_1b':
        G_1b=v
    if pos=='G_2b':
        G_2b=v
    if pos=='G_3b':
        G_3b=v
    if pos=='G_ss':
        G_ss=v
    if pos=='G_lf':
        G_lf=v
    if pos=='G_cf':
        G_cf=v
    if pos=='G_rf':
        G_rf=v        
```
